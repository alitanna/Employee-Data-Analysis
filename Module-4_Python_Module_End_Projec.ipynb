```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
file_path = "datasetforproject.xlsx"
df = pd.read_excel(file_path, engine="openpyxl")

# Display first few rows
df.head()
```

```python
# Replace 'Height' column with random values between 150 and 180
np.random.seed(42)
df['Height'] = np.random.randint(150, 181, size=len(df))

df.head()
```

```python
team_counts = df['Team'].value_counts()
team_percentages = (team_counts / len(df)) * 100

plt.figure(figsize=(12, 6))
sns.barplot(x=team_counts.index, y=team_counts.values, hue=team_counts.index, palette="viridis", legend=False)
plt.title("Employee Distribution Across Teams")
plt.xlabel("Team")
plt.ylabel("Number of Employees")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

team_percentages
```

```python
position_counts = df['Position'].value_counts()

plt.figure(figsize=(12, 6))
sns.barplot(x=position_counts.index, y=position_counts.values, hue=position_counts.index, palette="coolwarm", legend=False)
plt.title("Employee Distribution by Position")
plt.xlabel("Position")
plt.ylabel("Number of Employees")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
age_bins = [18, 25, 30, 35, 40, 50, 60]
age_labels = ['18-25', '26-30', '31-35', '36-40', '41-50', '51-60']
df['Age Group'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels, right=False)

age_group_counts = df['Age Group'].value_counts()

plt.figure(figsize=(10, 5))
sns.barplot(x=age_group_counts.index, y=age_group_counts.values, hue=age_group_counts.index, palette="magma", legend=False)
plt.title("Predominant Age Group")
plt.xlabel("Age Group")
plt.ylabel("Number of Employees")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

age_group_counts
```

```python
age_bins = [18, 25, 30, 35, 40, 50, 60]
age_labels = ['18-25', '26-30', '31-35', '36-40', '41-50', '51-60']
df['Age Group'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels, right=False)

age_group_counts = df['Age Group'].value_counts()

plt.figure(figsize=(10, 5))
sns.barplot(x=age_group_counts.index, y=age_group_counts.values, hue=age_group_counts.index, palette="magma", legend=False)
plt.title("Predominant Age Group")
plt.xlabel("Age Group")
plt.ylabel("Number of Employees")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

age_group_counts
```

```python
salary_by_team = df.groupby("Team")["Salary"].sum().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(x=salary_by_team.index, y=salary_by_team.values, hue=salary_by_team.index, palette="cividis", legend=False)
plt.title("Salary Expenditure by Team")
plt.xlabel("Team")
plt.ylabel("Total Salary")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
salary_by_position = df.groupby("Position")["Salary"].sum().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(x=salary_by_position.index, y=salary_by_position.values, hue=salary_by_position.index, palette="rocket", legend=False)
plt.title("Salary Expenditure by Position")
plt.xlabel("Position")
plt.ylabel("Total Salary")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['Age'], y=df['Salary'], color='blue', alpha=0.5)
sns.regplot(x=df['Age'], y=df['Salary'], scatter=False, color='red')
plt.title("Correlation Between Age and Salary")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.grid(True)
plt.tight_layout()
plt.show()

correlation = df['Age'].corr(df['Salary'])
print("Correlation between Age and Salary:", correlation)
```
